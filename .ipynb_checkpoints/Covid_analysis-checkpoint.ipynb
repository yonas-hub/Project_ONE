```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import warnings
warnings.filterwarnings('ignore')
```

```python
filename = "time_series_covid19.csv"
filedata=pd.read_csv(filename)
filedata
```

```python
#df.rename(columns={"A": "a", "B": "c"})
filedata=filedata.rename(columns={ "Country/Region":"country","4/23/20":"totalcases"})
filedata.dtypes
```

```python
filedata = filedata.loc[filedata["totalcases"]>7000]
filedata
```

```python
filedata.plot("country","totalcases")
plt.xticks(rotation=90)
```

```python
x=filedata["country"]
y=filedata["totalcases"]
# ax = plt.subplot(111)
# ax.bar(x, y, width=20)
# plt.xticks(rotation=90)
# plt.show()
plt.figure(figsize=(20,10))
plt.plot(x,y) ## This is your plot
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt;
import numpy as np
import matplotlib.pyplot as plt

objects = filedata["country"]
y_pos = np.arange(len(objects))
plt.figure(figsize=(20,10))
plt.bar(y_pos, filedata["totalcases"], align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.xlabel("Country")
plt.ylabel('total cases')
plt.title('total cases by country')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#plt.figure(figsize=(40,20))
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(30, 15))
N=len(filedata['totalcases'])
colors = np.random.rand(N)
ax1.scatter(filedata['country'], filedata['totalcases'],s=filedata["totalcases"]/1000,c=colors)
ax1.set_ylim()
plt.xticks(rotation=90)
ax2.bar(filedata['country'], filedata['totalcases'])
ax2.set_ylim()

# This only works for the last chart, even if we call it
# right after we create ax1!
plt.xticks(rotation=90);
```

```python
groupdata=filedata.groupby("country").sum()["totalcases"]
groupdata
```

```python
sorteddata=filedata.sort_values(by=['totalcases'],ascending=False)
sorteddata
```

```python
#filedata=filedata.sort_values(by=['totalcases'],ascending=False)
data=filedata.groupby('country').sum()['totalcases']
data=pd.DataFrame(data)
data=data.reset_index()
plt.figure(figsize=(40,20))
objects = data["country"]
y_pos = np.arange(len(objects))
plt.figure(figsize=(20,10))
plt.bar(y_pos, data["totalcases"], align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.xlabel("Country")
plt.ylabel('total cases')
plt.title('total cases by country')
plt.xticks(rotation=45, ha='right')
plt.show()
#data.plot.bar()
#data
```

```python
data=filedata.groupby('country').mean()['totalcases']
data=pd.DataFrame(data)
data=data.reset_index()
plt.figure(figsize=(50,20))
objects = data["country"]
y_pos = np.arange(len(objects))
size=data["totalcases"]/50
N=len(data['totalcases'])
colors = np.random.rand(N)
plt.figure(figsize=(20,10))
plt.scatter(y_pos, data["totalcases"], s=size,c=colors)
plt.xticks(y_pos, objects)
plt.xlabel("Country")
plt.ylabel('total cases')
plt.title('total cases by country')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
datedata=filedata[["country","1/22/20","2/22/20","3/22/20","4/22/20","totalcases"]]
datedata
```

```python
datedata["Jan-Feb"]=datedata["2/22/20"]-datedata["1/22/20"]
datedata["Feb-march"]=datedata["3/22/20"]-datedata["2/22/20"]
datedata["march-April"]=datedata["4/22/20"]-datedata["3/22/20"]
datedata=datedata.loc[datedata["totalcases"]>50000]
datedata
```

```python
datedata=datedata[["country","totalcases","Jan-Feb","Feb-march","march-April"]]
datedata=datedata.set_index(["country"])
datedata
```

```python
ax = datedata.plot.bar(x="totalcases",
                                      y=["Jan-Feb","Feb-march","march-April"],
                                      legend=True, figsize=(18,8))
for i, label in enumerate(list(datedata.index)):
    ax.annotate(label,(i, 30),ha="left",rotation=90)

plt.legend(loc = "upper right")
plt.title("New Covid cases in from Jan to April")
plt.ylabel("Covid cases")
plt.xlabel("Increase in from Jan - April")
plt.xticks(rotation=45);
```
